```python
# question1.a

from sklearn.decomposition import TruncatedSVD
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from numpy import linalg as LA
plt.style.use('classic')
img = Image.open('US_flag.jpeg')
from numpy.linalg import matrix_rank
```

```python
# convert image to grayscale
imggray = img.convert('LA')
imggray = img
# convert to numpy array
imgmat = np.array(list(imggray.getdata(band=0)), float)
# Reshape according to orginal image dimensions
imgmat.shape = (imggray.size[1], imggray.size[0])
plt.figure(figsize=(9, 6))
plt.imshow(imgmat, cmap='gray')
plt.show()
```

```python
#1.b

U, D, V = np.linalg.svd(imgmat)
```

```python
print("imgmat=",imgmat)
print("U=",U)
print("D=",D)
print("V",V)
```

```python
print(imgmat.shape)
print(U.shape)
print(D.shape)
print(V.shape)
```

```python
#1.c
s=imgmat.shape[0]
for k in range(1,s):
    Ak= np.dot(U[:,:k],np.dot(np.diag(D[:k]),V[:k,:]))
    print("Ak",k,)
    print(Ak)
```

```python
#1.d
def RelativeError(A,Ak):
    dif=np.subtract(A,Ak)
    error=(LA.norm(dif))/LA.norm(A)
    return error*100
```

```python
#1.e
E=[]
A=imgmat
count=0
for k in range(1,s):
    Ak= np.dot(U[:,:k],np.dot(np.diag(D[:k]),V[:k,:]))
    RE=RelativeError(A,Ak)
    E.append(RE)  
    if(RE<1 and count==0):
        count=count+1
        print(k)
```
