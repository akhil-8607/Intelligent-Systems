```python
from sklearn import preprocessing
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def MyPCD2D(X):
    preprocessing.StandardScaler().fit_transform(X)
    U, D, V = np.linalg.svd(X)
    T = V[:,:2]
    print(T)
    return(T)
```

```python
iris=datasets.load_iris()
X=iris.data[:,:4]
W=MyPCD2D(X)
```

```python
Y = np.dot(X,W)
```

```python
x_index = 0
y_index = 1

# this formatter will label the colorbar with the correct target names
formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)])

plt.figure(figsize=(5, 4))
plt.scatter(Y[:, 0], Y[:, 1], c=iris.target)
plt.xlabel(iris.feature_names[x_index])
plt.ylabel(iris.feature_names[y_index])

plt.tight_layout()
plt.show()
#clusters can be obersved in the 2D projection of the iris data set
```

```python
def MyPCA2D_transposed(XT):
    preprocessing.StandardScaler().fit_transform(X)
    U, D, V = np.linalg.svd(XT)
    T = U[:,:2]
    print(T)
    return(T)
```

```python
W1=MyPCA2D_transposed(X.T) 
```

```python
Y = np.dot(X,W1)
```

```python
x_index = 0
y_index = 1
formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)])
plt.figure(figsize=(5, 4))
plt.scatter(Y[:, 0], Y[:, 1], c=iris.target)
plt.xlabel(iris.feature_names[x_index])
plt.ylabel(iris.feature_names[y_index])

plt.tight_layout()
plt.show()
```

we are projecting the same data so the plot shape is the same for both the projection matrices. But the figure is reflection of one another since we are using transpose matrix to derive the projection matrix so there may be change of the sign in plotting the data.
